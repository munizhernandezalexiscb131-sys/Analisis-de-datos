```python
#setup
import numpy as np
import matplotlib.pyplot as plt

#generar datos
periodos = 25
tendencia = np.linspace(10, 35, periodos)
estacionalidad = 0 * np.sin(np.linspace(0, 2.6*np.pi, periodos))
ruido = np.random.normal(1.7, 2.1, periodos)   # ruido pequeño

serie = tendencia + estacionalidad + ruido

#graficacion
plt.plot(range(1, periodos + 1), serie, marker='o')
plt.title("Línea: serie temporal con tendencia y estacionalidad")
plt.xlabel("Periodo")
plt.ylabel("Valor")
plt.grid(True)
plt.show()

```
